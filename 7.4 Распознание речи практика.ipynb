### Google Text-To-Speech

[Ссылка на документацию](https://pypi.org/project/gTTS/)

```python
!pip install gtts
!pip install pydub
```

```python
from gtts import gTTS
import IPython.display as ipd
from pydub import AudioSegment
from pydub.playback import play

def synthesize_speech(text, lang='ru'):
    # Генерация речи из текста
    tts = gTTS(text=text, lang=lang, slow=False)
    tts.save("output.mp3")

    # Воспроизведение аудио
    audio = AudioSegment.from_mp3("output.mp3")
    play(audio)

    # Отображение аудио
    return ipd.Audio("output.mp3")

# Протестируйте функцию синтеза речи
text = "Привет, как дела? Это тест синтеза речи!"
synthesize_speech(text)
```

### Whisper

[Ссылка на документацию](https://github.com/openai/whisper/blob/main/README.md)

```python
# Установка необходимых библиотек
!pip install git+https://github.com/openai/whisper.git
!pip install setuptools rust
```

```python

# Импорт библиотек
import whisper
from google.colab import files

audio_file_ru = '/content/output.mp3'


# Загрузите модель Whisper
model = whisper.load_model("medium")

# Транскрипция аудио
result = model.transcribe(audio_file_ru, language="ru")
print(result["text"])
```

```python
audio_file_en = '/content/My_House.mp3'

result = model.transcribe(audio_file_en, language="en")
print(result["text"])
```

```python
# Транскрипция и перевод на английский
result_translation = model.transcribe(audio_file_ru, task="translate", language="en")
print("Перевод:")
print(result_translation["text"])
```

```python
# Включение временных меток
result_detailed = model.transcribe(audio_file_ru, word_timestamps=True)
print("Подробная информация о сегментах:")
for segment in result_detailed["segments"]:
   print(f"Начало: {segment['start']:.2f}, Конец: {segment['end']:.2f}, Текст: {segment['text']}")
```
