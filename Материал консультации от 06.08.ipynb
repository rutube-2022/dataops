```python
import json
import pandas as pd


with open('jobs.json', 'r') as file:
    data = json.load(file)

first_object = data[0]
pretty_first_object = json.dumps(first_object, indent=4)
print(pretty_first_object)
```

```python
df = pd.DataFrame(data)
pd.set_option('display.max_columns', 50)
df.head(10)
```

```python
df = df.dropna(axis=1)

df.head(10)
```

```python
df[['offerStockOrBonus', 'isFullRemote']] = df[['offerStockOrBonus', 'isFullRemote']].astype(int)

df
```

```python
language_size = df.groupby('language').size()

language_size
```

```python
for col in df.columns:
    if isinstance(df[col].iloc[0], list):
        print(f"Column '{col}' contains lists.")
```

```python
for col in df.columns:
    if isinstance(df[col].iloc[0], list):
        df[col] = df[col].apply(lambda x: str(x) if isinstance(x, list) else x)
```

```python
perkKeys_size = df.groupby('perkKeys').size()

perkKeys_size
```

```python
jobType_size = df.groupby('jobType').size()

jobType_size
```

```python
df.drop(columns=['_id', 'jobUrl','stateCategory', 'logoImg', 'cityCategory', 'postalCode', 'language', 'perkKeys', 'jobType'], inplace=True)

df.head(10)


```
