## Пример из биологии

### Принцип работы голосообразующего аппарата

Голосообразующий аппарат человека — это сложная система органов, которые работают вместе для создания звука и речи.

#### Основные компоненты голосообразующего аппарата
**Дыхательная система (респираторная часть)**
* **Легкие** это источник воздушного потока. Легкие обеспечивают подачу воздуха, необходимого для вибрации голосовых связок.
* **Дыхательные мышцы** это основные мышцы, участвующие в процессе дыхания, включают диафрагму и межреберные мышцы. Эти мышцы контролируют поступление и выдох воздуха, регулируя его объем и давление.

**Гортань (ларингеальная часть)**
* **Гортань** расположена в верхней части трахеи и содержит голосовые связки. Гортань выполняет важную функцию защиты дыхательных путей, а также участвует в фонации (создании звука).
* **Голосовые связки** это две эластичные связки мышечной ткани, расположенные внутри гортани. Когда они приводятся в напряжение и воздух проходит между ними, они начинают вибрировать, создавая звуковую волну.

**Резонаторы**
* **Глотка (фаринкс)** это полость между ротовой полостью и гортанью. Глотка участвует в усилении звуковых волн, создаваемых голосовыми связками.
* **Ротовая полость** это основное пространство для резонирования и модификации звука. Внутри ротовой полости расположены артикуляционные органы (язык, губы, зубы и мягкое небо), которые изменяют форму звукового сигнала.
* **Носовая полость** участвует в резонировании звуков, особенно носовых (таких как "м", "н").

**Артикуляционные органы**
(Язык, губы, зубы, мягкое и твердое небо) - эти органы участвуют в формировании различных звуков речи путем изменения формы и объема резонаторных полостей. Например, при произнесении звуков "т" и "д" язык касается альвеолярного гребня, а для звуков "м" и "н" мягкое небо опускается, открывая путь к носовой полости.


#### Механизм работы
* **Дыхание.** Голосообразование начинается с дыхания. Вдох обеспечивается сокращением диафрагмы и межреберных мышц, что увеличивает объем грудной клетки и позволяет воздуху поступить в легкие. При выдохе эти мышцы расслабляются, а воздух выходит из легких и проходит через гортань.

* **Фонация (создание звука)**. При выдохе воздух проходит через голосовые связки, расположенные в гортани. Когда голосовые связки находятся в напряжении и сближены, воздушный поток заставляет их вибрировать. Частота этих вибраций определяет высоту звука (например, высокий или низкий тон).

* **Резонирование.** Звуковые волны, созданные в гортани, проходят через резонаторные полости (глотку, ротовую и носовую полости). Эти полости усиливают звук и изменяют его характеристики (тембр). Размер и форма полостей, а также их способность изменяться под воздействием артикуляционных органов, играют ключевую роль в формировании различных звуков речи.

* **Артикуляция.** Последний этап — это артикуляция, то есть модификация звукового сигнала артикуляционными органами. Движения языка, губ, мягкого неба и челюсти изменяют поток воздуха и создают разнообразные звуки речи. Например, для звука "б" губы смыкаются, а для звука "с" язык поднимается к зубам.

### Принцип работы слуха

Ухо человека состоит из 3 частей:

**Наружное ухо**

Состоит из:
* **Ушной раковины**, которая собирает звуковые волны из окружающей среды и направляет их в слуховой проход.
* **Слухового прохода**, который проводит звуковые волны к барабанной перепонке, усиливая определенные частоты.

**Среднее ухо**

Состоит из:
* **Барабанной перепонки**, представляющей собой тонкую мембрану и которая отделяет наружное ухо от среднего уха. Она вибрирует в ответ на звуковые волны.
* **Слуховых косточек(молоточек, наковальня и стремечко)**, это 3 маленькие косточки, которые передают вибрации от барабанной перепонки к внутреннему уху. Эти косточки усиливают звуковые вибрации и передают их через овальное окно во внутреннее ухо.
* **Евстахиевой трубы** это канал, соединяющий среднее ухо с носоглоткой. Она уравновешивает давление воздуха на барабанную перепонку, что необходимо для правильного проведения звуковых волн.

**Внутреннее ухо**

Состоит из:

* **Улитки** это спирально закрученный орган, заполненный жидкостью. Улитка содержит орган слуха, называемый кортиев органом, который состоит из тысяч тонких волосковых клеток. Эти клетки преобразуют механические колебания в нервные импульсы.
* **Полукружных каналов** это три канала, расположенные перпендикулярно друг другу. Они не участвуют в слухе, но играют важную роль в поддержании равновесия тела.
* **Преддверие** это часть внутреннего уха, связанная с полукружными каналами и улиткой. Преддверие также участвует в поддержании равновесия.




#### Процесс слуха
* Сбор звука: Ушная раковина улавливает звуковые волны и направляет их в слуховой проход.
* Проведение звука: Звуковые волны достигают барабанной перепонки, которая начинает вибрировать.
* Усиление звука: Вибрации передаются на слуховые косточки среднего уха (молоточек, наковальня и стремечко), которые усиливают звуковой сигнал.
* Передача звука: Стремечко передает вибрации через овальное окно в улитку внутреннего уха.
* Преобразование звука в нервные сигналы: Вибрации вызывают движение жидкости внутри улитки, что приводит к изгибу волосковых клеток в кортиевом органе.
* Эти клетки преобразуют механические колебания в электрические импульсы.
* Передача нервных сигналов: Электрические импульсы передаются по слуховому нерву в мозг.
* Восприятие звука: Мозг обрабатывает полученные сигналы и распознает их как звуки.

## Распознавание речи (Speech-to-Text, S2T)

Распознавание речи (S2T) — это технология, которая
преобразует устную речь (аудиосигнал) в текст. Основная цель — точное определение и интерпретация слов, произнесенных человеком, для их последующего использования в цифровом формате.

**Задачи распознавания речи:**

* Преобразование устной речи в текст — создание транскрипции аудиосигнала, что позволяет пользователю взаимодействовать с системой с помощью голоса.
* Автоматизация и повышение эффективности — уменьшение необходимости ручного ввода текста, что актуально в таких сферах, как ведение заметок, расшифровка аудио и видео контента, заполнение документов и т.д.
* Улучшение пользовательского опыта — обеспечение удобного взаимодействия пользователей с устройствами и приложениями, особенно в условиях, когда использование клавиатуры и экрана затруднено.
* Доступность для людей с ограниченными возможностями — помощь людям с физическими ограничениями, которые не могут использовать традиционные средства ввода текста.
* Обработка естественного языка — улучшение диалоговых систем, чат-ботов, голосовых помощников и других приложений, требующих понимания естественного языка.
* Многоязычная поддержка и глобализация — возможность использования приложения людьми, говорящими на разных языках и диалектах, что важно для глобальных сервисов и международных компаний.


### Модули распознавания

#### Акустическая модель

Акустическая модель — это компонент системы распознавания речи, который преобразует звуковой сигнал (аудио) в вероятностные представления фонем (основных единиц звука в речи). Модель анализирует и распознает акустические признаки речи и связывает их с фонемами или словами, чтобы определить, что именно было произнесено.

**Основные функции**
* **Преобразование аудиосигнала в спектральное представление.**
Акустическая модель анализирует аудиосигнал, используя математические методы для преобразования его в спектрограмму, которая показывает, как меняются частоты звуков во времени.
Часто используются мел-частотные кепстральные коэффициенты (MFCC) или другие спектральные признаки, которые представляют особенности звукового сигнала.
* **Распознавание фонем.**
Модель анализирует спектральные признаки и сопоставляет их с фонемами — минимальными звуковыми единицами языка. Каждой фонеме присваивается вероятностное значение в зависимости от того, насколько хорошо она соответствует входному аудиосигналу.
Система работает на уровне кратких временных отрезков (фреймов), которые обычно имеют длительность 10-20 миллисекунд.
* **Предсказание вероятностей.**
Акустическая модель обучается предсказывать вероятность того, что определенный звуковой паттерн соответствует конкретной фонеме. Эти вероятности позже используются модулем декодирования для выбора наиболее вероятной последовательности слов.


>**Фонема** — *элементарная единица человеческой речи.*

##### Алгоритмы и технологии

###### **Скрытые марковские модели (Hidden Markov Models, HMM)**


Традиционный метод построения акустических моделей, где аудиосигнал рассматривается как последовательность наблюдений, а каждое наблюдение связано с определенной скрытой (непосредственно не наблюдаемой) фонемой.

Модель основана на вероятностях переходов между состояниями (фонемами) и вероятностях появления конкретных акустических признаков в этих состояниях.


**Пример работы на основе погоды.**

Давайте разберём HMM с помощью простого примера — представим, что мы пытаемся угадать погоду, зная только, что человек делает каждый день.

1. *Представьте, что у нас есть две вещи:*

**Скрытые состояния** это то, что мы хотим узнать, но не можем наблюдать напрямую. Например, "солнечно", "облачно" или "дождливо". Это и есть "скрытая" часть HMM.

**Наблюдения** это то, что мы можем увидеть или услышать. Например, мы знаем, что человек может заниматься одним из трех действий: "гулять", "делать покупки" или "сидеть дома". Мы можем видеть эти действия, но не знаем погоду напрямую.

2. *Переходы между состояниями*

В реальной жизни погода может меняться: если сегодня солнечно, завтра, возможно, будет облачно или останется солнечно, но, скорее всего, не станет дождливо. Скрытая марковская модель определяет вероятности переходов от одного состояния к другому.

Например, если сегодня "солнечно", есть 70% вероятность, что завтра тоже будет "солнечно", 20% вероятность, что станет "облачно", и 10% вероятность, что будет "дождливо".

3. *Вероятности наблюдений*

Каждое действие (наблюдение) имеет разную вероятность в зависимости от состояния (погоды). Например, если "солнечно", человек с 60% вероятностью будет "гулять", с 30% — "делать покупки" и с 10% — "сидеть дома".

Если же "дождливо", то вероятности будут другими: 10% для "гулять", 20% для "делать покупки" и 70% для "сидеть дома".

**Теперь давайте перенесем это на распознавание речи.**


**Скрытые состояния** в распознавании речи — это **фонемы**, или базовые звуковые элементы языка (например, звуки "к", "а", "т").

Эти состояния являются "скрытыми", потому что мы не можем напрямую наблюдать фонемы в аудиосигнале, мы можем только измерить акустические признаки (например, высоту звука, громкость и частоту).

**Наблюдения** — это **части аудиосигнала**, которые мы можем измерить, такие как спектральные коэффициенты (например, MFCC), представляющие частотные характеристики звука.

**Процесс распознавания речи с помощью HMM:**

* Шаг 1: Предобработка сигнала. Аудиозапись разбивается на короткие фрагменты (обычно 10-20 миллисекунд), для которых рассчитываются акустические признаки (например, спектральные коэффициенты).
* Шаг 2: Сопоставление наблюдений с состояниями. Каждому фрагменту сигнала присваивается вероятность того, что он соответствует определенной фонеме. Эта вероятность зависит от того, как похож данный фрагмент сигнала на "эталонные" звуковые паттерны фонем.
* Шаг 3: Использование вероятностей переходов между фонемами. Модель учитывает вероятности того, как одна фонема переходит в другую. Например, в русском языке звук "т" чаще всего переходит в звук "о" в словах типа "тот", чем в "щ" или "ю".
* Шаг 4: Поиск наилучшей последовательности. Используя алгоритм (например, алгоритм Витерби), модель находит наиболее вероятную последовательность фонем, которая могла бы соответствовать данному аудиосигналу.

###### **Глубокие нейронные сети**

Современные системы S2T используют глубокие нейронные сети для более точного распознавания речи. Нейронные сети позволяют учитывать сложные нелинейные зависимости между входными признаками и целевыми фонемами

DNN обучаются на огромных наборах данных, где каждому аудиосигналу соответствует его текстовая транскрипция. Это позволяет нейронным сетям идентифицировать сложные паттерны в аудиосигнале, которые коррелируют с определенными фонемами.



Как нейронная сеть "учится" распознавать речь?
1. **Обучение на больших объемах данных**

Чтобы сеть научилась распознавать речь, её "тренируют" на огромных наборах данных — миллионах записей реальной речи, которые уже были расшифрованы в текст. Например, тысячи часов записи людей, говорящих "Привет, как дела?" и другие фразы.
Сеть многократно "прослушивает" эти записи и учится сопоставлять звуковые паттерны с правильными текстами. Вначале она делает много ошибок, но благодаря обучению постепенно становится лучше.

2. **Корректировка весов**

В процессе обучения сеть настраивает свои внутренние параметры, которые называются весами. Каждый нейрон в сети имеет вес, который определяет, насколько сильно он "реагирует" на входной сигнал.
Когда сеть ошибается, она "понимает", какие веса были настроены неверно, и корректирует их с помощью математических алгоритмов, таких как градиентный спуск. Это похоже на то, как мы учимся: если мы понимаем, что ошиблись, мы пытаемся изменить свой подход, чтобы в следующий раз сделать лучше.

3. **Многократное повторение**

Этот процесс повторяется тысячи или миллионы раз, пока сеть не станет достаточно точной. Чем больше данных и времени на обучение, тем точнее становится сеть.

###### **Рекуррентные нейронные сети (RNN) и LSTM (Long Short-Term Memory)**

В обычной (прямой) нейронной сети каждый вход обрабатывается независимо от других, что хорошо для задач, где порядок данных не важен. Но в речи каждый звук и слово связаны с предыдущими: звук "к" в слове "кошка" зависит от звука "о", а "о" — от "ш", и так далее. Рекуррентные нейронные сети (RNN) решают эту проблему, запоминая предыдущие шаги и учитывая их при принятии решений.

RNN обучаются прогнозировать следующие элементы в последовательности, основываясь на предыдущих фреймах. Благодаря своей архитектуре с "петлями" внутри, RNN могут хранить информацию о предыдущих шагах, что позволяет модели учитывать контекст при прогнозировании текущего фрейма. Это свойство особенно полезно в распознавании речи, где произношение текущего звука (фонемы) зависит от того, какие звуки предшествовали ему.

**Как LSTM улучшает работу RNN:**

LSTM — это особый вид RNN, который был разработан для решения проблемы запоминания длительных зависимостей. Он включает в себя специальные механизмы, которые позволяют сети лучше запоминать информацию на протяжении долгого времени.

**Пример работы LSTM в распознавании речи:**

Давайте представим, как LSTM распознает фразу "Я иду в парк":

* **Входные данные.**

Сначала аудиосигнал разбивается на фрагменты (маленькие кусочки звука). Каждый фрагмент преобразуется в числовые данные — акустические признаки.

* **Обработка последовательности.**

LSTM начинает обрабатывать эти фрагменты один за другим. Например, когда он слышит звук, похожий на "Я", он сохраняет это в своей "ячейке памяти".
Затем, когда звучит "и", LSTM учитывает, что перед этим был "Я", и понимает, что это, скорее всего, часть той же фразы, а не отдельное слово.
Когда звучит "ду", LSTM помнит "Я" и "и", и "ду", что помогает ей сделать вывод о вероятности слова "иду".

* **Учет долгосрочных зависимостей.**

Весь процесс повторяется для остальных слов ("в парк"), и благодаря "воротам" LSTM запоминает долгосрочный контекст: "Я иду в парк". Так система понимает, что это цельная фраза, а не набор отдельных слов, и распознает её с высокой точностью.


###### **Трансформеры**

**Механизм внимания в распознавании речи**

Когда мы говорим, каждое слово или звук влияет на понимание всей фразы. Трансформеры анализируют каждый звук и слово и решают, насколько они важны для общего смысла.

Например, в предложении "Я иду в магазин", трансформер будет обращать внимание не только на каждое слово по отдельности, но и на их взаимосвязь. Так, слово "магазин" может сильно повлиять на значение всей фразы, и модель будет уделять ему больше внимания.

**Распознавание речи**

1. Преобразование аудиосигнала в числовые данные

Как и в других методах, аудиосигнал (звук) разбивается на короткие фрагменты. Каждый фрагмент анализируется и преобразуется в числовое представление, описывающее его акустические особенности.

2. Применение энкодера

Энкодер получает эти числовые данные и проходит через несколько слоев внимания. Каждый слой анализирует данные, фокусируясь на наиболее важных частях звука и учитывая как краткосрочные, так и долгосрочные зависимости.
В отличие от RNN или LSTM, трансформер может учитывать сразу всю фразу целиком. Это как если бы вы читали предложение и сразу понимали весь его смысл, а не обрабатывали каждое слово по отдельности.

3. Создание представлений

Энкодер создает представление для каждого фрагмента звука, которое учитывает его значимость и связь с другими фрагментами. Это представление передается в декодер.

4. Декодирование в текст

Декодер получает выходы энкодера и начинает генерировать текст. Он также использует механизм внимания, чтобы решить, какие звуки соответствуют каким буквам, фонемам или словам.
Декодер постепенно формирует текстовую версию произнесенной фразы, используя данные от энкодера.

5. Построение финального текста

Декодер заканчивает работу, когда вся фраза преобразована в текст. Таким образом, весь процесс становится гораздо быстрее и точнее по сравнению с традиционными методами.

#### Языковая модель

Языковая модель помогает компьютеру понять, какие слова и фразы чаще всего используются вместе в языке, и предсказать, какое слово будет следующим в предложении. Это делает взаимодействие с компьютером более естественным и понятным. Давайте разберемся, как это работает на простом примере.

Представьте, что вы читаете книгу, и на каждой странице есть предложения, где случайным образом исчезает одно слово. Если вы видите предложение "Я люблю ...", то, скорее всего, сможете догадаться, что пропущенное слово будет "кофе", "музыку" или что-то подобное. Вы делаете это, потому что у вас есть опыт и вы знаете, какие слова обычно следуют друг за другом.

Языковая модель делает то же самое, но с помощью компьютера. Она обучена на огромном количестве текстов (книг, статей, веб-страниц), чтобы научиться понимать, какие слова чаще всего встречаются вместе и какие вероятнее всего следуют за другими.

##### Принцип работы

1) Предсказание следующего слова:

Основная задача модели — предсказать, какое слово должно идти следующим в предложении. Например, если у нас есть начало предложения "Она пошла в...", модель попытается угадать следующее слово. На основе того, что она "прочитала" до этого, она может предложить такие слова, как "магазин", "школу" или "парк".

2) Использование N-грамм:

В традиционных системах использовался метод, называемый N-граммы. Этот метод анализирует, как часто определенные последовательности слов (например, пары или тройки слов) встречаются в тексте.
Например, биграмма (2-грамма) "доброе утро" встречается довольно часто, а триграмма (3-грамма) "доброе утро, дорогой" — реже, но все же имеет смысл в определенном контексте.
Чтобы модель могла делать точные предсказания, она должна быть обучена на большом количестве текстов, иначе ей будет сложно узнать, какие последовательности слов наиболее вероятны.

3) Ограничения N-грамм:

В методе N-грамм нужно выбирать небольшие значения N (например, 1, 2 или 3), чтобы модель могла эффективно работать, так как при большем количестве слов в последовательности предсказания становятся менее точными из-за нехватки данных.
Например, для модели униграмм (N=1) рассматривается только одно слово, для биграмм (N=2) — два слова подряд, а для триграмм (N=3) — три слова.

##### Как языковая модель улучшает распознавание речи

Когда человек говорит, слова часто могут быть произнесены нечетко или в шумной обстановке, что затрудняет их распознавание. Языковая модель помогает системе распознавания речи лучше понять, что было сказано, потому что она "знает", какие слова с большей вероятностью следуют друг за другом в определенном контексте.

Например, если компьютер услышит что-то вроде "Я люблю чит... книги", и слово "читать" было нечетко произнесено, языковая модель поможет угадать его по контексту. Она знает, что "чит..." перед словом "книги" скорее всего означает "читать", а не что-то другое.

##### Современные подходы к языковым моделям

1) Нейронные сети.

В отличие от традиционных N-грамм, нейронные сети, такие как рекуррентные нейронные сети (RNN) и трансформеры, способны учитывать контекст на гораздо более длинных отрезках текста. Они анализируют текст, понимая не только соседние слова, но и более широкую связь между ними.

2) Модели на базе трансформеров.

Эти модели используют продвинутый механизм внимания, который позволяет им одновременно учитывать все слова в предложении, понимая их значение в контексте.
Они могут анализировать более сложные структуры и связи в тексте, что делает их гораздо более точными и гибкими. Примеры сложных моделей: Whisper ASR от OpenAI, DeepSpeech, Wav2Vec.

### Области применения

Технология Speech-to-Text (STT) применяется в самых разных сферах, начиная от повседневной жизни и заканчивая специализированными профессиональными задачами. Одной из наиболее распространённых областей применения являются голосовые помощники и умные устройства, такие как Siri, Google Assistant, Amazon Alexa и Microsoft Cortana. Они используют STT для преобразования голосовых команд в текст, который затем интерпретируется системой для выполнения различных задач, включая управление устройствами умного дома, поиск информации в интернете, запуск приложений и выполнение заказов. Это позволяет пользователям управлять устройствами hands-free, делая технологии более доступными, особенно для людей с ограниченными возможностями.

В бизнес-среде технологии STT находят широкое применение в автоматизации бизнес-процессов, например, в контакт-центрах, автоматизированных системах звонков и CRM-системах. Здесь STT используется для распознавания и расшифровки телефонных разговоров и запросов клиентов, что позволяет автоматизировать процессы поддержки клиентов, мониторить качество обслуживания и создавать отчеты на основе разговоров. Это снижает затраты на персонал, повышает скорость обслуживания и улучшает качество клиентского сервиса.

STT также играет важную роль в медицине, где она используется для диктовки медицинских записей и автоматической транскрипции врачебных консультаций. Это помогает автоматизировать запись и хранение информации о пациентах, что значительно экономит время врачей, снижает риск ошибок и упрощает хранение и поиск информации. Такие приложения активно применяются для создания электронных медицинских карт, формирования отчетов и организации дистанционных консультаций.

Технологии STT способствуют повышению доступности и инклюзии, особенно для людей с нарушениями слуха или зрительными ограничениями. Они используются для создания приложений, которые помогают создавать текстовые версии аудиоинформации, тем самым делая контент доступным для большего числа пользователей. Примеры включают автоматическое создание субтитров для видеоконтента, текстовые сообщения в реальном времени и голосовые помощники для слепых пользователей.

В образовательной сфере STT активно используется для упрощения доступа к знаниям и улучшения восприятия учебного материала. Например, она помогает в автоматической транскрипции лекций, создании учебных материалов на основе записей, поддерживает интерактивные упражнения для изучения языков и создаёт субтитры для видеоуроков. Это особенно важно в условиях дистанционного обучения, когда студентам требуется доступ к материалам в разных форматах.

В медиа и индустрии развлечений STT помогает автоматизировать создание субтитров, транскрибировать подкасты и интервью, а также использовать голосовое управление в видеоиграх. Это улучшает доступ к контенту, увеличивает аудиторию и открывает новые возможности для взаимодействия с пользователями. STT также применяется в юридических и финансовых услугах для протоколирования встреч и автоматической транскрипции судебных заседаний, что помогает снизить затраты на персонал, ускорить документооборот и повысить точность записей.

В исследованиях и анализе данных STT используется для трансформации аудио в текстовый формат, что упрощает сбор и анализ данных в социологических и маркетинговых исследованиях. Она позволяет автоматизировать обработку данных из интервью, фокус-групп и семинаров, что сокращает время и затраты, а также повышает точность и эффективность исследований.

Наконец, технологии виртуальной и дополненной реальности (VR/AR) также активно используют STT для создания более естественного взаимодействия пользователя с виртуальными или дополненными мирами. Например, разрабатываются голосовые интерфейсы для VR-игр и приложений, а также реализуются виртуальные ассистенты, что улучшает пользовательский опыт и упрощает взаимодействие с виртуальными объектами.

## Синтез речи (Text-to-Speech, T2S)

Text-to-Speech — это технология, преобразующая написанный текст в естественную или синтезированную речь. Основная цель TTS — сделать текстовую информацию доступной в аудиоформате, позволяя пользователям "слушать" текст вместо его чтения. Это особенно полезно для людей с нарушениями зрения, для аудиокниг, голосовых ассистентов, а также в различных системах автоматизации.

### История развития

**Первые шаги в развитии TTS**

Ранние попытки синтеза речи были основаны на электронных схемах и механических системах. Один из первых известных проектов — "Voder" компании Bell Labs (1939), который воспроизводил отдельные звуки речи через набор механических и электронных компонентов.

Развитие формантных синтезаторов, таких как DECtalk, который стал одним из первых коммерчески успешных синтезаторов речи. Эти устройства использовали математические модели для воспроизведения основных звуков речи, но звучали довольно "металлически" и неестественно.

**Развитие в 1990-2000-х годах:**

Системы начали использовать записи реальных голосов для создания более естественного звучания. В основе метода лежит склейка коротких фрагментов записанной речи (дифонов) для формирования слов и предложений. Такие системы звучали лучше, но требовали значительных объемов данных для каждой новой голосовой модели.

Появление методов, основанных на скрытых марковских моделях (Hidden Markov Models, HMM). Эти модели позволяли генерировать речь на основе статистического анализа больших объемов текстовых и аудиоданных, что улучшило вариативность и плавность звучания.

**Современный этап**

Появление моделей на основе глубокого обучения (например, Tacotron от Google, WaveNet от DeepMind). Эти системы используют рекуррентные и сверточные нейронные сети для анализа и синтеза речи, что позволяет значительно улучшить качество синтезированной речи, делая её практически неотличимой от человеческой.

### Основные этапы синтеза речи

1) Анализ текста

На этом этапе происходит предварительная обработка текста для его подготовки к синтезу.

**Разбиение на предложения и слова:** Определение границ предложений и слов, учет пунктуации и других текстовых особенностей.

**Нормализация текста:** Преобразование аббревиатур, чисел, дат, специальных символов в их словесное представление (например, "12" преобразуется в "двенадцать").

**Лингвистический анализ:** Определение грамматической структуры предложения, включая части речи, иерархию слов и ударения.

**Произношение:** Преобразование текста в фонемы (набор звуков речи) с учетом особенностей языка и контекста. Здесь также учитываются такие элементы, как интонация, акценты и паузы.

2) Языковая модель

Языковая модель TTS анализирует и интерпретирует текст для выбора правильного произношения и интонации.

**Фонетическая транскрипция:** Преобразование слов в последовательность фонем. Важную роль играет транскрипция слов в зависимости от контекста и их использования.

**Прозодическая модель:** Определяет параметры интонации, тембра и ритма речи, чтобы синтезировать речь, которая звучит естественно и плавно.

3) Синтезатор речи

Синтезатор преобразует фонетическую транскрипцию в звуковую волну, которая становится синтезированным голосом.

Основные подходы к синтезу включают конкатенативный, формантный и нейросетевой методы, которые рассматриваются далее.

### Методы синтеза речи

#### Конкатенативный синтез

Конкатенативный синтез — это метод синтеза речи, который использует предварительно записанные фрагменты реальной человеческой речи (обычно называемые юнитами) для создания новой синтезированной речи. В основе метода лежит соединение (конкатенация) этих юнитов, чтобы формировать слова, фразы и предложения.

**Принцип работы**

* Метод предполагает наличие большого банка записанных звуковых фрагментов (юнитов), таких как отдельные фонемы, дифоны (переходы между двумя звуками), слоги, слова или даже целые фразы.
* Во время синтеза система выбирает и соединяет подходящие юниты из базы данных, чтобы создать аудиопоток, соответствующий входному тексту.
* Процесс включает в себя анализ текста, определение последовательности необходимых звуков и их соединение таким образом, чтобы синтезированная речь звучала как можно более естественно.


**На основе базы данных юнитов также работают:**

**Waveform Concatenation:** Этот метод предполагает прямое соединение (конкатенацию) звуковых сигналов, что позволяет сохранить естественные характеристики голоса и интонации.

**Unit Selection:** Более продвинутый подход, в котором используется алгоритм выбора оптимальных юнитов из большого набора записей. Алгоритм выбирает те фрагменты, которые минимизируют слышимые переходы и максимизируют естественность звучания.

#### Формантный синтез

Формантный синтез — это метод синтеза речи, который воспроизводит звуки человеческого голоса путем моделирования его акустических характеристик, таких как форманты. Форманты — это резонансные частоты, возникающие в акустическом тракте (ротовой и носовой полости) человека, которые определяют характерные звуковые свойства речи (особенно гласных звуков).

**Форманты** — это пиковые частоты в спектре звука, которые возникают из-за резонанса в голосовом тракте. Основные форманты (обычно F1, F2, и F3) используются для синтеза гласных звуков, а их частоты и амплитуды определяют акустические и артикуляторные свойства речи.

**F1:** Определяет высоту звука, связанную с открытостью рта. Например, звук [a] имеет более низкий F1, чем звук *[i]*.

**F2:** Определяет резонансные характеристики, связанные с местоположением языка (передний или задний ряд звуков).

**F3:** Влияет на такие аспекты, как качество звука, характерное для согласных и некоторых комбинаций звуков.


**Процесс синтеза речи с использованием формантного метода**

1) Анализ и обработка текста:

На этапе обработки текста система определяет структуру предложения, определяет фонемы и необходимые звуковые элементы, которые следует синтезировать.
Текст преобразуется в последовательность фонем, соответствующих звукам речи.

2) Генерация параметров формант:

Используя модели фонем, система генерирует параметры для каждой фонемы, включая основные частоты формант (F1, F2, F3 и выше), их амплитуды и длительность.
Эти параметры соответствуют различным звукам речи и варьируются в зависимости от особенностей произношения.

3) Моделирование голосового тракта:

Модель голосового тракта представляет собой акустическую модель, основанную на принципах фильтрации звука. Она включает в себя параметры длины и формы голосового тракта, которые могут изменяться в зависимости от звука, который нужно синтезировать.
Система использует математические модели, чтобы сгенерировать форму волны, которая имитирует звук, проходящий через голосовой тракт с определенными формантами.

4) Синтез звукового сигнала:

Генерируется основной сигнал — звук голоса, который модулируется формантами, чтобы получить целевую речевую форму. Этот сигнал представляет собой серию импульсов, генерируемых на основе частоты основной формы волны (вибрации голосовых связок).
Дополнительные алгоритмы сглаживания и фильтрации используются для улучшения качества синтезированной речи и устранения нежелательных шумов.

5) Постобработка:

Применяются методы сглаживания и коррекции, чтобы уменьшить "металлический" оттенок речи, который может возникнуть в результате математического моделирования.

#### Статистический параметрический синтез

Статистический параметрический синтез — это метод синтеза речи, в котором для моделирования акустических параметров речи (например, частотных характеристик, длительности и амплитуды) используются статистические модели, такие как скрытые марковские модели (Hidden Markov Models, HMM). Эти модели обучаются на больших объемах данных реальной речи и позволяют синтезировать речь, которая близко соответствует природной речи, но с более контролируемыми параметрами.

**HMM** — это вероятностные модели, которые описывают последовательности наблюдений (в данном случае акустических параметров) и переходы между различными состояниями (например, фонемами). Модели обучаются на большом объеме речевых данных и представляют вероятности переходов между состояниями и распределения наблюдений в каждом состоянии.

**Состояния HMM:** Каждая фонема в языке может быть представлена несколькими состояниями (например, начальное, среднее и конечное состояния звука).

**Вероятностные распределения:** В каждом состоянии моделируются такие параметры, как спектральные характеристики (мел-кепстральные коэффициенты), частоты основного тона (F0), энергии звука и длительность.

**Принцип работы:**

В отличие от методов, основанных на прямом использовании или комбинировании фрагментов записанной речи (как в конкатенативном синтезе), статистический параметрический синтез создает речь с нуля, используя статистические параметры, полученные из предварительно обученных моделей.

Система сначала обучается на больших объемах аудиоданных, анализируя их и строя статистические модели для каждого звука речи. Эти модели включают в себя параметры, такие как частоты формант, амплитуды, спектральные особенности и длительность звуков.

В процессе синтеза система использует обученные статистические модели для генерирования последовательностей параметров, из которых затем создается акустический сигнал с помощью синтезатора.

**Процесс синтеза речи с использованием HMM**

1) Подготовка и обучение модели

Собирается большой корпус записей речи с транскрипциями. Этот корпус используется для обучения моделей HMM.
Из каждой записи извлекаются акустические параметры (например, спектральные характеристики, частоты формант, F0 и т.д.), которые затем используются для обучения моделей.
Модели обучаются для каждого звука (фонемы) и контекста, учитывая переходы между состояниями и вероятности разных параметров для каждого звука в различных контекстах.

2) Анализ текста и языковое моделирование

Входной текст анализируется для определения последовательности фонем и их контекста.
Лингвистическая модель определяет вероятностное распределение акустических параметров, которые должны быть сгенерированы для каждой фонемы в зависимости от ее контекста (например, соседних звуков, ударения, положения в предложении и т.д.).

3) Генерация параметров речи

На основе входного текста и обученных HMM генерируются параметры, необходимые для синтеза речи, такие как спектральные параметры и F0.
Система оптимизирует сгенерированные параметры с помощью специальных алгоритмов (например, алгоритмов динамического программирования), чтобы улучшить качество и естественность речи.

4) Синтез акустического сигнала

Сгенерированные параметры передаются на вход синтезатору, который создает звуковой сигнал.
Используются фильтры и методы сглаживания, чтобы устранить возможные шумы и артефакты и сделать звучание речи более естественным.

5) Постобработка

Применяются дополнительные методы для улучшения качества звука, такие как корректировка интонации, усиление или ослабление звуков, фильтрация шумов.

#### Нейронные сети и глубокое обучение в синтезе речи

Использование нейронных сетей и методов глубокого обучения в синтезе речи (Text-to-Speech, TTS) подразумевает создание моделей, которые могут автоматически обучаться воспроизводить человеческую речь на основе текстовых данных. Эти модели обучаются на больших наборах данных записанной речи и текстов, чтобы научиться связывать написанные слова с их звуковым выражением.

**Основные архитектуры синтеза речи**

**Tacotron**

Это архитектура нейронной сети, предложенная Google, которая использует последовательную модель на основе рекуррентных нейронных сетей (RNN) для преобразования текста в мел-спектрограмму, которая затем преобразуется в звуковой сигнал.

**Tacotron 2**

Улучшенная версия, использующая комбинацию рекуррентных нейронных сетей и сверточных сетей для более точного синтеза. Вместо использования традиционного вокодера, как в первой версии, Tacotron 2 использует WaveNet для синтеза высококачественного звука из мел-спектрограммы.

**WaveNet**

WaveNet — это глубокая сверточная нейронная сеть, разработанная DeepMind, которая генерирует речь на уровне выборок (samples), предсказывая каждую следующую выборку звуковой волны с высокой частотой дискретизации.

**FastSpeech**

Это архитектура, предназначенная для повышения скорости и стабильности синтеза речи по сравнению с Tacotron и WaveNet. Она использует полностью сверточную архитектуру, заменяя рекуррентные сети на сверточные для улучшения скорости обработки.

**FastSpeech 2**

Улучшенная версия, которая дополнительно моделирует такие аспекты, как громкость, продолжительность и основная частота звука (F0), чтобы еще лучше контролировать выходную речь.

**VITS**

VITS — это новая архитектура, использующая вариационные автоэнкодеры (VAE) и нормализующие потоки (Normalizing Flows) для обучения высококачественным распределениям данных речи.


### Области применения

Конвертация текста в речь (TTS) применяется в различных отраслях и сферах деятельности, помогая улучшить доступность, персонализацию и эффективность взаимодействия с пользователями.

В образовании TTS помогает сделать обучение более доступным и удобным. Например, технология используется для создания аудиоуроков и лекций, особенно в тех случаях, когда доступ к преподавателям ограничен или недоступен. Это особенно полезно для дистанционного обучения, позволяя студентам прослушивать учебные материалы в удобное время. TTS также применяется в мобильных приложениях для изучения языков, создавая произношение слов и фраз, что улучшает качество взаимодействия и помогает практиковать аудирование и произношение. В образовательных системах чтения текста TTS может быть использован для автоматического чтения текстов, что особенно важно для студентов с нарушениями зрения или дислексией, помогая улучшить их грамотность и понимание прочитанного. В целом, технология играет важную роль в обеспечении равного доступа к образованию для людей с ограниченными возможностями, позволяя использовать текстовые материалы в аудиоформате.

В индустрии развлечений TTS активно применяется для создания аудиокниг и подкастов, что значительно сокращает затраты на производство контента. Благодаря этому издатели и авторы могут быстро и экономически эффективно переводить текстовые произведения в аудиоформат. Современные TTS-системы позволяют настраивать голос, скорость чтения и интонацию, что делает прослушивание более персонализированным. В видеоиграх и интерактивных развлечениях TTS используется для озвучивания второстепенных персонажей и создания динамических диалогов, что позволяет разработчикам создавать разнообразные игровые сценарии без необходимости записи множества дублей с актерами. Также TTS помогает генерировать озвучку для различных игровых ситуаций, создавая более глубокий и динамичный опыт для игрока. В мультимедиа и кино TTS может использоваться для автоматической генерации субтитров и озвучивания контента на разных языках, что делает его доступным для большего числа зрителей.

В индустрии обслуживания TTS-технологии играют важную роль в голосовых помощниках и умных устройствах. Например, такие голосовые ассистенты, как Siri, Alexa и Google Assistant, используют TTS для озвучивания ответов на запросы, уведомлений, напоминаний и другой информации. Технология также интегрируется в умные домашние устройства, такие как колонки, телевизоры и даже холодильники, обеспечивая голосовое управление и доступ к информации. В автомобильных системах TTS используется для озвучивания навигационных инструкций и других сообщений, что делает вождение более безопасным. TTS также широко применяется в автоматизации колл-центров, где используется для озвучивания интерактивных голосовых меню и предоставления клиентам информации в автоматическом режиме, что снижает нагрузку на операторов и повышает качество обслуживания. Технология помогает автоматизировать обработку стандартных запросов, например, предоставление информации о балансе счета или статусе заказа, а также отправку уведомлений, таких как напоминания о встречах или изменениях в расписании.

В медицине TTS имеет множество применений, поддерживая людей с ограниченными возможностями и улучшая их качество жизни. Для людей с нарушениями зрения технология позволяет озвучивать тексты, электронные книги, веб-страницы и интерфейсы мобильных приложений, обеспечивая доступ к информации. Это повышает их независимость и автономность, позволяя взаимодействовать с миром через слуховые каналы восприятия. TTS также используется в устройствах для альтернативной и дополнительной коммуникации (AAC), таких как Tobii Dynavox или Proloquo2Go, которые позволяют людям с нарушениями речи общаться с помощью синтезированной речи. Это особенно полезно для пациентов с афазией, боковым амиотрофическим склерозом (БАС) или церебральным параличом. В телемедицине и удаленном мониторинге TTS применяется для автоматизированных напоминаний о приеме лекарств, графике посещений врача или выполнении лечебных процедур, а также для предоставления психологической поддержки через расслабляющие аудиозаписи и медитации. Люди с когнитивными нарушениями могут также использовать TTS для управления временем и выполнения ежедневных задач с помощью систем напоминаний, а также для улучшения когнитивных функций через озвучивание инструкций и упражнений в процессе реабилитации.
